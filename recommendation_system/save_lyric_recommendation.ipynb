```python
# 한국어만 추출하는 함수
def extract_korean_lyrics(text):
    # 정규 표현식을 사용하여 한국어만 추출
    korean_only = re.sub(r'[a-zA-Z]+', ' ', text)
    return korean_only



# 한국어 텍스트 전처리 함수 (예: 띄어쓰기 정리)
def preprocess_korean_lyrics(text):
    # 특수 문자 제거 & 양 옆 공백 제거
    text = re.sub(r'[^\w\s]', '', text)
    text = text.strip()

    # 형태소 분석하기 -> 형태소를 분석하되, 어간까지 변환하도록 하여 최대한 간결하게 표현하도록 함
    okt = Okt()
    tokens = okt.morphs(text, stem = True)

    # 불용어 제거
    stopwords = ['의','가','이','은','들','는','좀','을', '만','에서', '에게', '이며', '걍','과','도','를','으로','자','에','와','한','하다']
    tokenized_sentence = [word for word in tokens if word not in stopwords]

    # # 토큰화된 결과를 문자열로 결합
    # processed_text = ' '.join(tokens)

    return tokenized_sentence     # 리스트로 반환



def preprocess_english_lyrics(text):
    # 영어 텍스트를 소문자로 변환
    english_text = text.lower()

    # punctuation 제거
    translator = str.maketrans('', '', string.punctuation)
    english_text = english_text.translate(translator)

    #tokens = word_tokenize(english_text)
    tokens = WordPunctTokenizer().tokenize(english_text)

    # stopwords 제거
    stopwords = set(nltk.corpus.stopwords.words('english'))
    tokes_nostopwords = [token for token in tokens if token not in stopwords] # 정의한 불용어 제거

    lemmatized_eng_words = []
    for token in tokes_nostopwords:
      lemmatized_eng_words.append(nltk.WordNetLemmatizer().lemmatize(token))

    return lemmatized_eng_words   # 리스트로 반환




# 가사 데이터를 처리하는 함수
def process_lyrics(lyrics):

    if re.search(r'[a-zA-Z]', lyrics) and re.search(r'[가-힣]', lyrics):    # 영어와 한국어 포함하는 경우
        korean_lyrics = extract_korean_lyrics(lyrics)
        preprocessed_korean_lyrics = preprocess_korean_lyrics(korean_lyrics)
        return preprocessed_korean_lyrics

    elif re.search(r'[가-힣]', lyrics) and not re.search(r'[a-zA-Z]', lyrics):
        preprocessed_korean_lyrics = preprocess_korean_lyrics(lyrics)
        return preprocessed_korean_lyrics


    elif re.search(r'[a-zA-Z]', lyrics) and not re.search(r'[가-힣]', lyrics):
        preprocessed_english_lyrics = preprocess_english_lyrics(lyrics)
        return preprocessed_english_lyrics



# 가사가 들어 있는 "가사" 칼럼에 적용합니다.
lyric_processing(df)
df['Processed_Lyrics'] = df['New_lyrics'].apply(process_lyrics)
df
```

```python
from sklearn.model_selection import train_test_split as sklearn_train_test_split
from sklearn.metrics.pairwise import cosine_similarity

def train_test_split(playlist):
  # Split each playlist into train (80%) and test (20%) sets
    train, test = sklearn_train_test_split(playlist, test_size=0.2, random_state=42)
    return train, test

def vectorize_lyrics(lyric, model):
    """Vectorize a lyric by averaging all its word vectors."""
    words = lyric.split()
    vectors = [model.wv[word] for word in words if word in model.wv.index_to_key]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

def song_recommend(playlist, model, num_recommendations=20):
    # Vectorize the lyrics of the songs in the user playlist using the 'Lyrics' column
    user_playlist_vectors = playlist['New_lyrics'].apply(lambda x: vectorize_lyrics(x, model)).tolist()
    mean_user_vector = np.mean(user_playlist_vectors, axis=0)  # Calculate the mean vector of the user playlist

    # Vectorize all song lyrics
    df['vector'] = df['New_lyrics'].apply(lambda x: vectorize_lyrics(x, model))

    # Calculate cosine similarity between the mean user vector and all song vectors
    df['similarity'] = df['vector'].apply(lambda x: cosine_similarity([x], [mean_user_vector])[0][0])

    # Exclude songs that are already in the user playlist
    playlist_song_ids = playlist['SongID'].tolist()
    recommended_songs = df[~df['SongID'].isin(playlist_song_ids)]

    # Sort the songs by similarity to recommend
    recommended_songs = recommended_songs.sort_values(by='similarity', ascending=False).head(num_recommendations)
    return recommended_songs

# 숨겨진 곡들과 추천된 곡들을 기반으로 NDCG를 계산
def compute_ndcg(recommended_items, hidden_items):
    dcg = 0.0
    idcg = 0.0
    for i, rec_item in enumerate(recommended_items):
        if rec_item in hidden_items:
            dcg += 1.0 / np.log2(i + 2)
    for i in range(len(hidden_items)):
        idcg += 1.0 / np.log2(i + 2)
    return dcg / idcg
```

```python
lyric_processing(playlist_1)
train_song, test_song = train_test_split(playlist_1)
recommended_song = song_recommend(train_song, lyric_model)

recommended_song[['SongID', 'Title', 'Singer', 'Genre']]
```

```python
# Define a function to vectorize lyrics using the trained Word2Vec model
def vectorize_lyrics_lyric_model(lyrics, model):
    vectorized_lyrics = [model.wv[word] for word in lyrics if word in model.wv]
    return np.mean(vectorized_lyrics, axis=0) if vectorized_lyrics else np.zeros(model.vector_size)

# Function to recommend songs based on a given playlist
def recommend_songs_lyric_model(user_playlist, all_songs, model, num_recommendations=20):
    user_playlist_vectors = user_playlist['Processed_Lyrics'].apply(lambda x: vectorize_lyrics_lyric_model(x, model)).tolist()
    mean_user_vector = np.mean(user_playlist_vectors, axis=0)
    all_songs['vector'] = all_songs['Processed_Lyrics'].apply(lambda x: vectorize_lyrics_lyric_model(x, model))
    all_songs['similarity'] = all_songs['vector'].apply(lambda x: cosine_similarity([x], [mean_user_vector])[0][0])
    song_ids_in_playlist = set(user_playlist['SongID'])
    recommendations = all_songs[~all_songs['SongID'].isin(song_ids_in_playlist)]
    recommended_songs = recommendations.sort_values(by='similarity', ascending=False).head(num_recommendations)
    return recommended_songs
```
