```python
# 파일 경로
csv_files = [
    "/content/drive/MyDrive/Github/playlist/playlist_1.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_2.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_3.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_4.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_5.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_6.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_7.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_8.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_9.csv",
    "/content/drive/MyDrive/Github/playlist/playlist_10.csv",

]
```

```python
# 좋아요 수, 댓글 수 number type으로 변환
def clean_number(x):
    return int(x.replace(',', '').replace('개', '').strip())
```

```python
# 파일 로드 및 초기 전처리
all_data = []
for file in csv_files:
    df = pd.read_csv(file)
    df['Likes'] = df['Likes'].apply(clean_number) # number type으로 변환
    df['CommentsCount'] = df['CommentsCount'].apply(clean_number) # number type으로 변환
    all_data.append(df)

# 모든 데이터프레임을 하나로 합치기
combined = pd.concat(all_data, ignore_index=True)
```

```python
# 일단 모든 데이터 병합
combined
```

```python
from collections import Counter
import ast

# 문자열을 리스트로 변환하기
tags_as_lists = [ast.literal_eval(tag_str) for tag_str in combined['Tags']]

# 모든 태그를 하나의 리스트로 만들기
all_tags = [tag for sublist in tags_as_lists for tag in sublist]

# 태그 빈도 계산하기
tag_counts = Counter(all_tags)

# 결과 출력하기
for tag, count in tag_counts.items():
    print(f"Tag: {tag}, Count: {count}")
```

```python
# 제거할 태그들
remove_tags = ['#힙합', '#발라드', '#댄스댄스', '#인디', '#록', '#메탈', '#이무진', '#트로트', '#성인가요', '#트롯맨', '#RnB', '#국내RnB', '#댄스']
# 문자열을 리스트로 변환하고 제거할 태그들을 제거
new_tags = [str([tag for tag in ast.literal_eval(tag_str) if tag not in remove_tags]) for tag_str in combined['Tags']]

# 업데이트된 태그로 data 딕셔너리 업데이트
combined['NewTag'] = new_tags

# 결과 확인
print(combined['NewTag'])
```

```python
import ast

def convert_string_to_list(s):
    try:
        return ast.literal_eval(s)
    except:
        return []
```

```python
# SongID를 기준으로 그룹화하고 필요한 처리 수행
def aggregate_rows(group):
    # 'Rank' 평균 처리
    rank_avg = group['Rank'].mean()

    # 'Likes'와 'CommentsCount'의 최대값 처리
    max_likes = group['Likes'].max()
    max_comments = group['CommentsCount'].max()

    # 문자열로 저장된 리스트를 실제 리스트로 변환
    group['NewTag'] = group['NewTag'].apply(convert_string_to_list)

    # 'NewTag' 값을 합치기 위해 중복 제거 후 리스트로 변환
    combined_tags = list(set(tag for tags_list in group['NewTag'] for tag in tags_list))



    # 첫번째 행을 기반으로 새로운 행을 만들되, 'Rank', 'Likes', 'CommentsCount'만 수정
    new_row = group.iloc[0].copy()
    new_row['Rank'] = rank_avg
    new_row['Likes'] = max_likes
    new_row['CommentsCount'] = max_comments
    new_row['NewTag'] = combined_tags

    return new_row
```

```python
# SongID가 같을 경우 하나의 데이터로 처리
# Rank -> 같은 노래가 있는 경우 두 노래의 Rank값의 평균값으로 변환
# Likes, CommentsCount -> 같은 노래가 있는 경우, 좋아요 수와 댓글 수가 더 큰 값으로 값 할당
final_df = combined.groupby('SongID').apply(aggregate_rows).reset_index(drop=True)
final_df
```

```python
# 'Tags' 컬럼 제거
new_df = final_df.drop(columns='Tags')
```

```python
import pandas as pd
# 파일 경로
file_name = "/content/drive/MyDrive/Github/csv_files/final_all_songs.csv"

df = pd.read_csv(file_name)
```

```python
def recommend_songs(songs, user_tags):
    matching_songs = songs[songs['NewTag'].apply(lambda tags: any(tag in tags for tag in user_tags))]   # 초기추천: matching songs라는 DF새로 생성하기

    if user_tags:
        matching_songs['tagmatches'] = matching_songs['NewTag'].apply(lambda tags: sum(tag in tags for tag in user_tags))
        matching_songs = matching_songs.sort_values(by='tagmatches', ascending=False)

    return matching_songs

# 사용자 태그 입력
#user_input = input("원하는 태그를 쉼표로 구분하여 입력하세요: ")
user_tags = ["#새벽", "#휴식", "명곡"]

# 추천 결과 출력
recommended = recommend_songs(df, user_tags)
recommended
```
