```python
file_name1 = "/content/drive/MyDrive/Github/RecommenderSystem/data/all_song_with_song_sentence.csv"
file_name2 = "/content/drive/MyDrive/Github/RecommenderSystem/data/all_songs_with_processed_lyric.csv"
file_name3 = "/content/drive/MyDrive/Github/RecommenderSystem/data/songid_emotions.csv"
```

```python
import pandas as pd
song_sentence = pd.read_csv(file_name1, encoding='utf-8')
lyric = pd.read_csv(file_name2, encoding='utf-8')
emotion = pd.read_csv(file_name3, encoding='utf-8')
```

```python
merged_data = lyric.merge(emotion, on='SongID', how='left')
merged_data = merged_data.merge(song_sentence[['SongID', 'era', 'seasons', 'song_sentence']], on='SongID', how='left')
```

```python
merged_data
```

```python
# Replace commas with spaces in the 'song_sentence' column
merged_data['song_sentence'] = merged_data['song_sentence'].str.replace(',', ' ')

# Display the first few rows of the updated 'song_sentence' column
merged_data[['song_sentence']]
```

```python
# Recreate the song_sentence
merged_data['song_sentence'] = merged_data['song_sentence'] + ' ' + \
                              merged_data['emotions']
```

```python
# Replace commas with spaces in the 'song_sentence' column
merged_data['song_sentence'] = merged_data['song_sentence'].str.replace(',', '')
```

```python
merged_data[['song_sentence']]
```

```python
# 새로운 파일로 저장
merged_data.to_csv("/content/drive/MyDrive/Github/RecommenderSystem/data/final_songs.csv", index=False, encoding='utf-8')
```
