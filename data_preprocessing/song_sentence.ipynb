```python
df['ReleaseDate'] = pd.to_datetime(df['ReleaseDate'])
df.dtypes
```

```python
df['era'] = df['ReleaseDate'].dt.year.apply(lambda x: f"{(x // 10) * 10}")
df['era']
```

```python
df.head(3)
```

```python
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'
```

```python
df['seasons'] = df['ReleaseDate'].dt.month.apply(get_season)
df.head(2)
```

```python
#데이터프레임 다루는 법
#첫번째: 이름으로 호출하기.
#두번째: iloc/loc로 호출하기. <- 인덱스, 이름으로 slice하는 방법

df1 = df[['SongID', 'Singer', 'Composers', 'Lyricists', 'Arrangers']]
df1.head(2)
```

```python
import re
```

```python
def name_cleansing(text):
  text = str(text)
  cleansed = re.sub(r'\([^)]*\)', '', text)
  return re.sub(r'\s', '', cleansed)  # 모든 공백 제거
```

```python
df1['Singer'].apply(name_cleansing)
```

```python
df1['Singer_1'] = df1['Singer'].apply(name_cleansing)
df1
```

```python
df1['Singer_1'].isna().unique()
```

```python
df1['Composers_1'] = df1['Composers'].apply(name_cleansing)
df1['Lyricists_1'] = df1['Lyricists'].apply(name_cleansing)
df1['Arrangers_1'] = df1['Arrangers'].apply(name_cleansing)
df1
```

```python
df[['Singer', 'Composers', 'Lyricists', 'Arrangers']] = df1[['Singer_1', 'Composers_1', 'Lyricists_1', 'Arrangers_1']]
df.head()
```

# nan 처리(도현)
#### final_all_songs : 원본

#### selected_data : nan이 있는 행 인덱스에 해당하는 행렬만 str.replace('nan', '')

#### final_all_songs_cleaned : 모든 행렬에서 str.replace('nan', '')


```python
import pandas as pd

# pre_songs_final.csv 파일 불러오기.
df = pd.read_csv("/content/drive/MyDrive/capstone/pre_songs_final.csv", encoding='utf-8')
```

```python
# 행 값에 null이 있는 경우, 해당하는 열의 이름과 null값인 행의 인덱스 나열
null_rows = df[df.isnull().any(axis=1)]
for column in null_rows.columns:
    null_indices = null_rows[column][null_rows[column].isnull()].index.tolist()
    if null_indices:
        print(f'열 "{column}" : 행 "{null_indices}"')
```

```python
final_all_songs = df
final_all_songs.head(1)
```

```python
def data_cleansing(playlist):
  # 작곡가가 null이면 가수로 채우기
  playlist['Composers'].fillna(playlist['Singer'], inplace=True)

  # 작사가가 null이면 작곡가로 채우기
  playlist['Lyricists'].fillna(playlist['Composers'], inplace=True)

  # 편곡가가 null이면 작곡가로 채우기
  playlist['Arrangers'].fillna(playlist['Composers'], inplace=True)

#playlist['ReleaseDate'].astype(str) + " " + \
def create_song_sentence(playlist):
  # Transforming the user's playlist in the same way
  playlist['song_sentence'] = playlist['Singer'].astype(str) + " " + \
                         playlist['Genre'].astype(str) + " " + \
                         playlist['Composers'].astype(str) + " " + \
                         playlist['Lyricists'].astype(str) + " " + \
                         playlist['Arrangers'].astype(str) + " " + \
                         playlist['era'].astype(str) + " " + \
                         playlist['seasons'].astype(str)
```

```python
create_song_sentence(final_all_songs)
```

```python
final_all_songs[['Title', 'song_sentence']].tail()
```

```python
# nan이 있는 행 인덱스
selected_indices = [1, 2, 3, 4, 8, 18, 30, 70, 71, 104, 133, 152, 153, 159, 175, 182]

# nan이 있는 행 인덱스에 해당하는 데이터프레임 추출
selected_data = final_all_songs.loc[selected_indices, ['Title', 'song_sentence']]

# nan이 있는 행에 해당하는 행렬만 확인
selected_data[['Title', 'song_sentence']]
```

```python
# 'nan' 값을 빈 문자열로 대체
# nan이 있는 행 인덱스만 추출한 행렬에서 nan값이 안 보이게 함 = selected_data
selected_data['song_sentence'] = selected_data['song_sentence'].str.replace('nan', '')
selected_data[['Title', 'song_sentence']]
```

```python
# 'nan' 값을 빈 문자열로 대체
# 모든 행렬에서 nan값이 안 보이게 함 = final_all_songs_cleaned
final_all_songs_cleaned = final_all_songs.copy()
final_all_songs_cleaned['song_sentence'] = final_all_songs_cleaned['song_sentence'].str.replace('nan', '')
final_all_songs_cleaned[['Title', 'song_sentence']]
```

# 2pre_songs_final.csv 파일 저장

```python
#2pre_songs_final.csv : final_all_songs.csv에 'era', 'seasons', 'song_sentence'열이 추가됨.
# pre_songs_final.csv : final_all_songs.csv에 'era'열과 'seasons'열이 추가됨.


# 'song_sentence'열을 pre_songs_final.csv에 추가
df['song_sentence'] = final_all_songs_cleaned['song_sentence']

# 새로운 파일로 저장
df.to_csv("/content/drive/MyDrive/capstone/csv_file/2pre_songs_final.csv", index=False, encoding='utf-8')
```
